```python
#import the necessary librRIES
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator

```

```python
dataset_path="PetImages"
```

```python
datagen=ImageDataGenerator(
    #Normalize pixels
    rescale=1./255,
    #data augmentation
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    #split dataset
    validation_split=0.2
)
```

```python
# LOad training Data
#Training data
train_data=datagen.flow_from_directory(
    dataset_path,
    target_size=(128,128),
    batch_size=32,
    class_mode='binary',
    subset='training'
)
```

```python
# LOad training Data
#Training data
val_data=datagen.flow_from_directory(
    dataset_path,
    target_size=(128,128),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
print(train_data.class_indices)
```

```python
model=keras.Sequential()
# 1.Simple Images ---> Few Layers
 
# ex: handwritten numbers
 
# (28*28)
 
# 2 conc 2 pool
 
# 2.Medium Complexity
 
# Cats vs Dogs
 
# (128*128)
 
# (224*224)
 
# 3 Conv + 3 Pool.
 
 
 
```

```python
#first convolutional layer
model.add(
    keras.layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu',
        input_shape=(128,128,3)
    )
)
```

```python
#first pooling layer
model.add(
    keras.layers.MaxPooling2D(
        pool_size=(2,2)


    )
)
```

```python
#second convolutional layer
model.add(
    keras.layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu'
    )
)
```

```python
#second pooling layer
model.add(
    keras.layers.MaxPooling2D(
        pool_size=(2,2)
    )
)
```

```python
model.add(keras.layers.Flatten())
```

```python
model.add(
    keras.layers.Dense(units=128,activation='relu')
)
```

```python
model.add(
    keras.layers.Dense(
        units=1,
        activation='sigmoid'
    )
)
```

```python
#View model architecture
model.summary()
```

```python
#Compile CNN
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(train_data,epochs=5,validation_data=val_data)
```

```python
loss,accuracy=model.evaluate(val_data)
print("accuracy:",accuracy)
```
